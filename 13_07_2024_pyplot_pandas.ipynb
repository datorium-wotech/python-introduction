```python
# FROM SYNTHETIC DATA
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import PathPatch
from matplotlib.path import Path

N = 400
t = np.linspace(0, 2 * np.pi, N)
r = 0.5 + np.cos(t)
x, y = r * np.cos(t), r * np.sin(t)

fig, ax = plt.subplots()
ax.plot(x, y, "k")
ax.set(aspect=1)
```

```python
# Sample data representing the average temperature (in degrees Celsius) from 2010 to 2014
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
df
```

```python
# VERSION ONE

plt.plot(df['Year'], df['Avg_Temperature'], marker='o') #first value is X, second is Y

plt.title('Average temperature from 2010 to 2014') #we added title
plt.xlabel('Year') #title on x axis
plt.ylabel('Temperature') #title on y axis

plt.xticks(df['Year']) #year values are rounded
plt.grid(True)
plt.show() #show you the visualization
```

```python
# VERSION TWO

df.plot(x='Year', y='Avg_Temperature', kind='line') #we are using pandas
plt.title('Average temperature from 2010 to 2014') #we added title
plt.xlabel('Year') #title on x axis
plt.ylabel('Temperature') #title on y axis

plt.xticks(df['Year']) #year values are rounded
plt.grid(True)
plt.show() #show you the visualization
```

The difference between `df.plot()` and `plt.plot()` lies in the library-specific functions used and the level of control and customization each offers. Here are the details:

1. `df.plot()`

This is a high-level plotting function provided by the `pandas` library, which acts as a wrapper around `matplotlib`. It simplifies the process of creating a plot directly from a `pandas` DataFrame. (When we say that df.plot() in pandas is a "wrapper around matplotlib," it means that the pandas plotting functions are built on top of matplotlib.)

**Pros:**
- **Simplicity**: Easy to use, especially for quick plots directly from a DataFrame.
- **Integration**: Works directly with DataFrame columns, making it very convenient for exploratory data analysis.
- **Less Code**: Requires less code to generate a plot.

**Cons:**
- **Less Customization**: While it's great for quick plots, it offers less fine-grained control over the plot appearance compared to `matplotlib`.

2. `plt.plot()`

This is a low-level plotting function from the `matplotlib` library. It requires more code but offers extensive customization options for creating detailed and complex plots.

**Pros:**
- **Customization**: Allows for extensive customization of all aspects of the plot (titles, labels, colors, markers, grid, etc.).
- **Flexibility**: Suitable for creating complex and highly customized plots.
- **Control**: Provides more control over the plotting process and appearance.

**Cons:**
- **Complexity**: Requires more lines of code and a deeper understanding of `matplotlib`.

Summary
- **`df.plot()`** is convenient for quick, simple plots directly from a DataFrame with minimal code.
- **`plt.plot()`** offers more detailed and customizable plotting options but requires more code and understanding of the `matplotlib` library.

Both methods are useful depending on the context: use `df.plot()` for quick visualization and `plt.plot()` for more detailed and customized plots.
Code Abstraction
By abstracting the complexity of matplotlib, pandas plotting functions make it easier to create standard plots quickly. For users who need basic visualizations, this means they can focus on their data rather than learning the intricacies of matplotlib.
Customization with Matplotlib
While pandas provides a simplified interface, the plots created are still matplotlib objects. This means you can further customize these plots using matplotlib functions if needed.

```python
# BAR CHART

plt.bar(df['Year'], df['Avg_Temperature'], color='skyblue') #color can be selected by you
plt.title('Avg temperature from 2010 to 2014')
plt.xlabel('Year')
plt.ylabel('Temperature')
```

```python
# H-BAR CHART

plt.barh(df['Year'], df['Avg_Temperature'], color='orange') #color can be selected by you
plt.title('Avg temperature from 2010 to 2014')
plt.xlabel('Temperature')
plt.ylabel('Year')
```

```python
# HISTOGRAM - presents the distribution of a signle variable, frequency, represents numerical data

data2 = {'Scores': [85, 92, 78, 20, 89, 90, 82, 56, 60, 23, 90, 40, 77]}
df2 = pd.DataFrame(data2)
df2

plt.hist(df2['Scores'], bins=5, color='green', edgecolor='black')
```

```python
# CUSTOM BIN EDGES
bin_edges = [0, 40, 60, 80, 90, 100]
plt.hist(df2['Scores'], bins=bin_edges, color='green', edgecolor='black')
```

```python
# SCATTER PLOT - used to visualize the relationship between two variables

data3 = {'X': [1,2,3,5,5,2], 'Y':[5,8,7,4,9,5]}
df3 = pd.DataFrame(data3)

plt.scatter(x = df3['X'], y=df3['Y'])
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
# PIE CHART - represents parts of a whole

data4 = {'Category': ['A', 'B', 'C', 'D'], 'Values': [15, 30, 25, 40]}
df4 = pd.DataFrame(data4)
df4

plt.pie(df4['Values'], labels = df4['Category'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# BOX PLOT - distribution of a dataset and identifies outliers
plt.boxplot(df4['Values'])
plt.grid(True)
plt.show()
```
